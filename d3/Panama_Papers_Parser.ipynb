```python
import pandas as pd
import numpy as np
import json
```

```python
# COUNTRY TO COUNTRY RELATIONSHIPS
with open('edges_by_country_20131231_GBR.json') as json_data:
    d = json.load(json_data)
    
df_p = pd.DataFrame(d)

country1 = np.array(df_p.country_code_1, dtype=object)

date = np.array(df_p.date, dtype=object)

country2 = np.array(df_p.country_code_2, dtype=object)

PD = pd.crosstab(country1, [date, country2], rownames=['country1'], colnames=['date', 'country2'])

dftest = pd.DataFrame(PD)
df_bool = dftest.loc[:] > 0

rows = len(df_bool)
cols = len(df_bool.columns)

print "["
for r in range(0,rows):
    results = []
    for c in range(0,cols):
        if df_bool.columns.values[c][1] == "NULL":
            continue
        if df_bool.iloc[r,c] == True:
            results.append(df_bool.columns.values[c][1])

    ans = (','.join('"' + item + '"' for item in results))
    print '{\"name\":\"',df_bool.index[r], '\",\"imports\":[',ans,']},'
    
print "]"
```

```python
# ENTITY TO ENTITY RELATIONSHIPS
with open('edges_by_entity_20131231_10161058.json') as json_data:
    d = json.load(json_data)
    
df_p = pd.DataFrame(d)

node1 = np.array(df_p.node1_name, dtype=object)

date = np.array(df_p.date, dtype=object)

node2 = np.array(df_p.node2_name, dtype=object)

PD = pd.crosstab(node1, [date, node2], rownames=['node1'], colnames=['date', 'node2'])

dftest = pd.DataFrame(PD)
df_bool = dftest.loc[:] > 0

rows = len(df_bool)
cols = len(df_bool.columns)

print "["
for r in range(0,rows):
    results = []
    for c in range(0,cols):
        if df_bool.columns.values[c][1] == "NULL":
            continue
        if df_bool.iloc[r,c] == True:
            results.append(df_bool.columns.values[c][1])

    ans = (','.join('"' + item + '"' for item in results))
    print '{\"name\":\"',df_bool.index[r], '\",\"imports\":[',ans,']},'
    
print "]"
```

```python
# Entities represented in columns but not rows. Needed for d3 vis.
rows = len(df_bool)
cols = len(df_bool.columns)
col_names = []
row_names = []

for c in range(0,cols):
    col_names.append( df_bool.columns.values[c][1])
    
for r in range(0,rows):
    row_names.append( df_bool.index[r])

    
diff = list(set(col_names) - set(row_names))
for i in range(0,len(diff)):
    print '{\"name\":\"',diff[i], '\",\"imports\":[]},'
```
